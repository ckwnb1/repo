```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, regexp_extract,count,when
import pandas as pd
# 创建 SparkSession
spark = SparkSession.builder.appName("OlympicsEDA").getOrCreate()
# 定义文件路径
csv_file_path = "E:\\hku\\cloud cluster\\ex4\\after1958.csv"


semicolon_csv_file_path = "E:\\hku\\cloud cluster\\ex4\\after1958_semicolon.csv"

# 使用 Pandas 读取原始 CSV 文件，并将其转换为分号分隔符
df = pd.read_csv(csv_file_path, encoding="UTF-8")
df.to_csv(semicolon_csv_file_path, sep=';', index=False, encoding="UTF-8")
# 读取使用分号分隔符的新 CSV 文件
spark_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("encoding", "UTF-8") \
    .option("delimiter", ";") \
    .csv(semicolon_csv_file_path)

# 打印 Schema
spark_df.printSchema()

# 显示数据的前几行
spark_df.show(50)

# 获取数值列的统计信息aa
spark_df.describe().show()

# 计算总行数
total_rows = spark_df.count()
print(f"Total Rows: {total_rows}")

# 打印 DataFrame 的列名和数据类型
print("DataFrame 的列名和数据类型:")
print(spark_df.dtypes)


```

```python
# 定义正则表达式模式来匹配 "aaaa Summer" 格式
pattern = r"^\d{4} Summer$"

# 使用正则表达式提取并过滤符合格式的行
invalid_games_df = spark_df.filter(regexp_extract(col("Games"), pattern, 0) == "")

# 显示不符合格式的行
invalid_games_count = invalid_games_df.count()

if invalid_games_count > 0:
    print(f"以下行的 'Games' 列不符合 'aaaa Summer' 格式:")
    invalid_games_df.show(invalid_games_count, truncate=False)
else:
    print("所有行的 'Games' 列均符合 'aaaa Summer' 格式.")
```

```python
# 统计 Games 列每一个元素对应的 Weight 列为 NULL 的行和 Height 列为 NULL 的行
na_stats = spark_df.groupBy("Games").agg(
    count(when(col("Weight").isNull(), True)).alias("Weight_NA_Count"),
    count(when(col("Height").isNull(), True)).alias("Height_NA_Count")
)



# 按 Weight_NA_Count 列进行降序排序
na_stats_sorted = na_stats.orderBy(col("Weight_NA_Count").desc(), col("Height_NA_Count").desc())

# 显示排序后的统计结果
na_stats_sorted.show()
```

```python
# 计算总行数
total_rows = spark_df.count()

# 计算每列缺失值的数目和百分比
missing_data_stats = spark_df.select(
    [(count(when(col(c).isNull(), c)) / total_rows).alias(c + "_missing_percentage") for c in spark_df.columns]
)

# 显示缺失数据的百分比
missing_data_stats.show(truncate=False)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, when, avg

# 创建 SparkSession
spark = SparkSession.builder.appName("OlympicsEDA").getOrCreate()

# 读取使用分号分隔符的新 CSV 文件
semicolon_csv_file_path = "E:\\hku\\cloud cluster\\ex4\\after1958_semicolon.csv"
spark_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("encoding", "UTF-8") \
    .option("delimiter", ";") \
    .csv(semicolon_csv_file_path)

# 计算总行数
total_rows = spark_df.count()

# 计算每列缺失值的数目和百分比
missing_data_stats = spark_df.select(
    [(count(when(col(c).isNull(), c)) / total_rows).alias(c + "_missing_percentage") for c in spark_df.columns]
)

# 显示缺失数据的百分比
missing_data_stats.show(truncate=False)

# 计算 Age、Height 和 Weight 列的平均值
avg_values = spark_df.select(
    avg(col("Age")).alias("avg_age"),
    avg(col("Height")).alias("avg_height"),
    avg(col("Weight")).alias("avg_weight")
).collect()[0]

avg_age = avg_values["avg_age"]
avg_height = avg_values["avg_height"]
avg_weight = avg_values["avg_weight"]

# 用平均值填充 Age、Height 和 Weight 列中的 null 值
spark_df = spark_df.fillna({"Age": avg_age, "Height": avg_height, "Weight": avg_weight})

# 验证填充是否成功
spark_df.select("Age", "Height", "Weight").show(5)
```

```python
# 计算总行数
total_rows = spark_df.count()

# 计算每列缺失值的数目和百分比
missing_data_stats = spark_df.select(
    [(count(when(col(c).isNull(), c)) / total_rows).alias(c + "_missing_percentage") for c in spark_df.columns]
)

# 显示缺失数据的百分比
missing_data_stats.show(truncate=False)
```

```python
output_csv_file_path = "E:\\hku\\cloud cluster\\ex4\\after1958_filled2.csv"
# 将 Spark DataFrame 写入一个目录
temp_output_path = "E:\\hku\\cloud cluster\\ex4\\after1958_filled_temp"
spark_df.write \
    .option("header", "true") \
    .option("delimiter", ",") \
    .mode("overwrite") \
    .csv(temp_output_path)

# 合并文件
import shutil
from pathlib import Path

# 读取临时目录中的所有 part 文件并合并到一个文件中
output_file = Path(output_csv_file_path)
with open(output_file, 'wb') as wfd:
    for part_file in Path(temp_output_path).rglob('part-*'):
        with open(part_file, 'rb') as fd:
            shutil.copyfileobj(fd, wfd)

# 删除临时目录
shutil.rmtree(temp_output_path)
```
